# Univariate:Single variable analysis

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('tips.csv')
df.head()
```

## i.Catagorical Data

```python
#countplot()
sns.countplot(x='day',data=df)
df['day'].value_counts()
```

```python
#piechart
df['day'].value_counts().plot(kind='pie',autopct='%.2f')
```

## ii.Numerical Data

```python
#histogram
plt.hist(df['tip'])
plt.show()
```

```python
#distplot()
sns.distplot(df['tip'])
```

```python
#displot()
sns.displot(df['tip'])
```

```python
#boxplot()
sns.boxplot(df['size'])
```
