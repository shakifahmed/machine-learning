# Bivariate & Multivariate:two or multi variable analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = pd.read_csv('tips.csv')
tips.head()
```

```python
#scatter plot(Numerical-Numerical)
sns.scatterplot(x='total_bill',y='tip',data=tips)#bivariate
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='sex',style='smoker',size='size')#multivariate
```

```python
titanic = sns.load_dataset('titanic')
```

```python
#barplot(Numerical-Catagorical)
sns.barplot(x='pclass',y='fare',data=titanic) #bivariate
sns.barplot(x='pclass',y='fare',data=titanic,hue='sex') #multivariate
```

```python
#distplot(Numerical-Catagorical)
sns.distplot(titanic[titanic['survived']==0]['age'])
sns.distplot(titanic[titanic['survived']==1]['age'])
```

```python
#boxplot(Numerical-Catagorical)
sns.boxplot(x='pclass',y='age',hue='sex',data=titanic)
```

```python
flights = sns.load_dataset('flights')
```

```python
#heatmap(Catagorical-Catagorical)
pivot_tab = flights.pivot(values='passengers',index='month',columns='year')
sns.heatmap(pivot_tab)
```

```python
sns.heatmap(pd.crosstab(titanic['pclass'],titanic['sex']))
```

```python
# sns.heatmap(titanic.groupby('pclass')['survived'].mean()*100)
pd.crosstab(titanic['pclass'],titanic['sex'])
(titanic.groupby('pclass')['survived'].mean()*100).plot(kind='bar')
```

```python
iris = sns.load_dataset('iris')
```

```python
#pairplot(Numeric)
sns.pairplot(iris,hue='species')
```

```python
#lineplot(Numerical-Numerical)
sns.lineplot(flights.groupby('year')['passengers'].sum())
```
