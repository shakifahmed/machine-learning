# ➥ Feature Construction

```python
import pandas as pd
```

```python
titanic = pd.read_csv('train.csv')
titanic.head()
titanic = titanic[['Survived','SibSp','Parch']]
titanic
```

```python
titanic['Family_size'] = titanic['SibSp'] + titanic['Parch']

def myfunc(num):
    if num == 0:
        return 0
    elif num <= 4:
        return 1
    else:
        return 2
titanic['Family_type'] = titanic['Family_size'].apply(myfunc)  
```

```python
titanic.drop(['SibSp','Parch'],inplace=True,axis=1)
```
