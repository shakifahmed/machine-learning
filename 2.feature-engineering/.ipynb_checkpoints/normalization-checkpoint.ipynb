# Normalization Scaler

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
wine = pd.read_csv('wine_data.csv',header=None,usecols=[0,1,2]) # if there is no header in dataframe
wine.columns = ['class label','alcohol','malic acid']
wine
```

### Split Datasets

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(wine.drop('class label',axis='columns'),
                                                 wine['class label'],test_size=0.3,random_state=0)
```

```python
X_train.shape,X_test.shape
```

## MinMax Scaler

### Step1
**MinMax Scaler**

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
Scaler = MinMaxScaler()
Scaler.fit(X_train)
```

### Step2
**Transformation**

```python
X_train_scaled = Scaler.transform(X_train)
X_test_scaled = Scaler.transform(X_test)
```

### Step3
**Array to DataFrame**

```python
X_train_scaled = pd.DataFrame(X_train_scaled,columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled,columns=X_test.columns)
```

### Step5
**Effect of Normalization**

```python
#kde plot
fig, (x1, x2) = plt.subplots(1,2, figsize = (15, 5))
sns.kdeplot(X_train['alcohol'],ax=x1)
sns.kdeplot(X_train['malic acid'],ax=x1)
x1.set_title('Before Scalling')

sns.kdeplot(X_train_scaled['alcohol'],ax=x2)
sns.kdeplot(X_train_scaled['malic acid'],ax=x2)
x2.set_title('After Scalling')
```

```python
#compare individulally(before & after scalling has no difference)
#for age
fig, (x1, x2) = plt.subplots(1,2, figsize = (15, 5))
sns.kdeplot(X_train['alcohol'],ax=x1)
x1.set_title('Before Scalling')
sns.kdeplot(X_train_scaled['alcohol'],ax=x2)
x2.set_title('After Scalling')
#for EstimatedSalary
fig, (x3, x4) = plt.subplots(1,2, figsize = (15, 5))
sns.kdeplot(X_train['malic acid'],color='r',ax=x3)
x3.set_title('Before Scalling')
sns.kdeplot(X_train_scaled['malic acid'],color='r',ax=x4)
x4.set_title('After Scalling')
```
