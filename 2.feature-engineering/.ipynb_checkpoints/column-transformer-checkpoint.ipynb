# Column Transformer**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
covid = pd.read_csv('covid_toy.csv')
covid.head()
```

```python
covid.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
X_train,X_test,y_train,y_test = train_test_split(covid.iloc[:,0:5],
                                                    covid.iloc[:,-1],
                                                    test_size=0.2,
                                                    random_state=0)
```

```python
tf = ColumnTransformer(transformers=[
    ('f1',SimpleImputer(),['fever']),
    ('f2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('f3',OneHotEncoder(sparse_output=False,drop='first'),['gender','city'])],
    remainder='passthrough')
```

```python
tf
```

```python
tf.fit_transform(X_train)
```
