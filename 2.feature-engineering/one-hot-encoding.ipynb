# One Hot Encoding
**Applied on Nominal(when the categorical variable has no order or ranking among its values) Dataset**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

```python
cars = pd.read_csv('cars.csv')
cars.head()
```

```python
cars = cars.iloc[:,2:]
cars.head()
```

```python
cars['fuel'].value_counts()
```

```python
cars['owner'].value_counts()
```

```python
X_train,X_test,y_train,y_test = train_test_split(cars.iloc[:,0:2],
                                                    cars.iloc[:,-1],
                                                    test_size=0.2,
                                                    random_state=0)
```

```python
X_train
```

```python
ohe = OneHotEncoder(sparse_output=False,dtype='int32') #sparse_output=False otherwise it don't give o/p in array format
```

```python
X_train_scaled = ohe.fit_transform(X_train)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled,dtype='int32')
```
